```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data_5.odt", comments='#')
```

```python
time = data[:,0]
mx = data[:,1]

```

```python
plt.figure()
plt.plot(time,mx)
```

```python
import scipy.fftpack

psd = (np.abs(scipy.fftpack.fft(mx)))
f_axis = scipy.fftpack.fftfreq(55526, d=11e-9/55526)

plt.plot(f_axis/1e9, psd)
plt.xlim([27.5, 37.5])
plt.ylim([0, 0.3e-6])
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
plt.grid()
```

```python
d_psd = np.diff(psd)/np.diff(f_axis)

plt.plot(f_axis[1:len(f_axis)]/1e9, d_psd)
plt.xlim([27.0, 37.5])
plt.ylim([-0.25e-15, 0.25e-15])
plt.grid()
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
```

```python
data = np.loadtxt("data_6_fig_6.odt", comments='#')
time = data[:,0]
mx = data[:,1]
plt.figure()
plt.plot(time,mx)
```

```python
import scipy.fftpack

import scipy.interpolate



psd = (np.abs(scipy.fftpack.fft(mx)))
f_axis = scipy.fftpack.fftfreq(19308, d=9.65e-9/19308)

plt.plot(f_axis/1e9, psd)
plt.xlim([32, 45])
plt.ylim([0, 0.3e-6])
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')
plt.grid()


x = f_axis[:5000]
y = psd[:5000]

xnew = np.linspace(x[0], x[4999], 100000) 

a_BSpline = scipy.interpolate.make_interp_spline(x, y)
y_new = a_BSpline(xnew)

plt.figure
plt.plot(xnew/1e9, y_new)
plt.xlim([33, 38])
plt.ylim([0, 1.5e-7])
plt.show()
```

```python
d_psd = np.diff(y_new)/np.diff(xnew)



plt.plot(xnew[:99999]/1e9, d_psd)
plt.xlim([33, 38])
plt.ylim([-1.3e-15, 1.3e-15])
plt.grid()
plt.xlabel('f (GHz)')
plt.ylabel('Psa (a.u.)')

```
